```python
import numpy as np
import matplotlib.pyplot as plt
```

## Function for plotting and calculating paths

```python
def generate_noise_arr(num_path, T):
    # seed ensures we use the same noise each time
    np.random.seed(0)
    
    # we want num_path draws of size T each
    noise_list = np.random.normal(0, 1, size=(num_path, T))
    
    return noise_list
```

```python
def generate_ar_path(phi, T):
    # generates 5 noise sequences
    noise_arr = generate_noise_arr(5, T)
    
    # generates 5 AR paths
    path_list = []    
    for n in noise_arr:
        path = [n[0]]
        for i in range(1, noise_arr.shape[1]):
            path.append(path[i - 1] * phi + n[i])
        
        path_list.append(path)
        
    return(np.array(path_list))
```

```python
def print_stats_and_plot(phi, T):
    ar_path_list = generate_ar_path(phi, T)
    
    # plot all paths together
    plt.plot(ar_path_list.T)
    
    mean_list = []
    std_list = []
    for ar_path in ar_path_list:
        mean_list.append(ar_path.mean())
        std_list.append(ar_path.std())
    
    print('PHI:', phi, ' |  T:', T)
    print('Sample means:', ar_path_list.mean(axis=1))
    print('Sample standard deviations:', ar_path_list.std(axis=1))
```

## Sample size list

```python
T_LIST = [100, 1000, 10000]
```

## $\phi$ = 0.9

```python
phi = 0.9
```

```python
print_stats_and_plot(phi, T_LIST[0])
```
